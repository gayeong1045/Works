# 문제1-7) '지역고용통계' 를 위한 전처리 함수를 작성하세요.

````python
def get_processing(path) :
    raw_employment = pd.read_excel( path, sheet_name = '데이터', skiprows=1 )
    sel_rows = raw_employment.index[raw_employment['산업별']=='전산업']
    employment_data = raw_employment.loc[sel_rows]
    employment_data['전체종사자'] = pd.to_numeric(employment_data['전체종사자'])
    emp_data = employment_data[['지역별', '전체종사자']]
    emp_data.columns = ['지역명', '고용자수']
    
    def get_juso(x) :
        temp = x.split(' ')
        sido = temp[0]
        if len(sido) != 4 : 
            return (sido[:2])
        else :
            return (sido[0]+sido[2])
        return 
    
    emp_data['지역명'].dropna().apply(get_juso)
    emp_data['시도'] = emp_data['지역명'].dropna().apply(get_juso) 
    emp_data['구군'] = emp_data['지역명'].dropna().apply(lambda x : x.split(' ')[1])
    
    return emp_data   
````
